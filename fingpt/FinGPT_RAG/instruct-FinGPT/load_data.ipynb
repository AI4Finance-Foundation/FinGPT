```python
import numpy as np

# 生成一个随机的浮点型参数
x = np.random.normal(size=10)

# 确定数据范围
min_x, max_x = np.min(x), np.max(x)
scale = (max_x - min_x) / 255
zero_point = np.round((0 - min_x) / scale)

# 线性量化
x_quantized = np.round(x / scale) + zero_point

# 逆量化
x_dequantized = (x_quantized - zero_point) * scale

# 打印结果
print("Original data:\n", x)
print("Quantized data:\n", x_quantized)
print("Dequantized data:\n", x_dequantized)
```
